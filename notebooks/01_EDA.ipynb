# 📊 Exploratory Data Analysis (EDA)
This notebook explores the Pima Indians Diabetes Dataset to understand distributions, correlations, and relationships before building ML models.

### 1. Load and inspect `models/diabetes_data.csv`
First, we load the dataset and take a look at the data structure and types.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('../models/diabetes_data.csv')

# Display first few rows and info
display(df.head())
print(df.info())
```

### 2. Class distribution bar chart
Understanding the target class distribution helps us know if the dataset is imbalanced. Imbalanced datasets might require SMOTE or different class weights.

```
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Outcome', palette='Set2')
plt.title('Class Distribution (0: No Diabetes, 1: Diabetes)')
plt.show()
```

### 3. Feature histograms for all 8 original features
We look at the distribution of each feature to spot skewness or outliers.

```
original_features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
df[original_features].hist(figsize=(12, 10), bins=20, color='skyblue', edgecolor='black')
plt.tight_layout()
plt.show()
```

### 4. Correlation heatmap
Heatmaps help us identify highly correlated features which might cause multicollinearity, or see which features correlate most with the outcome.

```
plt.figure(figsize=(10, 8))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()
```

### 5. Box plots: each feature vs Outcome
Box plots illustrate how feature values differ between the Diabetic and Non-Diabetic groups, highlighting potential predictive power.

```
plt.figure(figsize=(15, 10))
for i, col in enumerate(original_features, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(data=df, x='Outcome', y=col, palette='Set2')
    plt.title(f'{col} by Outcome')
plt.tight_layout()
plt.show()
```

### 6. Pairplot of top 4 features colored by Outcome
Visualizing pairwise relationships among the top correlated features gives insight into how they interact linearly or non-linearly to separate the classes.

```
top_features = ['Glucose', 'BMI', 'Age', 'DiabetesPedigreeFunction', 'Outcome']
sns.pairplot(df[top_features], hue='Outcome', palette='Set2', diag_kind='kde')
plt.show()
```

### 7. Summary of key findings
- **Imbalance**: The dataset has more Non-Diabetic (Class 0) than Diabetic (Class 1) cases.
- **Glucose**: The most critical indicator; higher glucose strongly correlates with diabetes.
- **BMI & Age**: Also show strong separation between classes.
- **Correlations**: Glucose, BMI, and Age have the highest linear correlation with the Outcome. Engineered features (like GlucoseBMI) should capture the interaction well.
